Task 2

Whether factors like age,Years at the company or Monthly income have role in determining if people will leave or not.

Create a scatterplot with age of the employees v/s Attrition

Drawing scatter plot withAge Vs Attrition

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
sns.scatterplot(x="Age",y="Attrition",data=data,hue="Attrition",style="Attrition",size="Attrition",sizes=(180,120))
plt.grid(color='k',linestyle='--',linewidth=0.5)
```

Create bar chart to show  realation between years at company and Attrition

```python
plt.figure(figsize=(11,6))
sns.histplot(data=data,x='YearsAtCompany',y="Attrition",color='b',edgecolor='linen',alpha=0.7,bins=10)
plt.title("Histogram of YearsAtCompany Vs Attrition")
plt.xlabel("YearsAtCompany")
plt.ylabel("Attrition")
plt.show()
```

Attrition decreases as the number of years in the company increases.

Create a histogramshowing realationship between monthly income and Attrition

```python
plt.figure(figsize=(11,6))
sns.histplot(data=data,x='MonthlyIncome',y="Attrition",color='b',edgecolor='linen',alpha=0.7,bins=10)
plt.title("Histogram of MonthlyIncome Vs Attrition")
plt.xlabel("MonthlyIncome")
plt.ylabel("Attrition")
plt.show()
```

We can see that, attrition is low at high salaries
